# GARCH Model of volatility

```python
ticker = ['^GSPC']
# ticker = ['GOGL']
start="2010-01-01"
end="2019-12-31"
last_obs="2017-01-01"
start_obs="2017-01-02"
```

```python
sp=st.asset(ticker,10000,start,end)
sp.get_data()
```

```python
garch = arch_model(sp.returns, vol='GARCH', p=1, q=1, dist='normal')
fgarch = garch.fit(disp='off') 
resid = fgarch.resid
st_resid = np.divide(resid, fgarch.conditional_volatility)

fgarch.summary()
```

```python
tickers = ['AAPL','META', 'GOOGL', 'AMZN']
weights = np.array([.25, .3, .15, .3])
initial_investment = 1000000

start="2018-01-01"
estimation="2021-12-31"
startobs='2022-01-01'
end=dt.date.today()
```

```python
port_2=pt.portfolio(tickers,weights,initial_investment,start,estimation)
port_2.get_data()
```

```python
port_2.returns
```

# n day montecarlo

```python
n=5
q=divmod(len(actual),n)[0]
r=q*n
actual=actual.iloc[:r,]
a=np.zeros(q)
for i in range(0,q,n):
    a[i]=np.sum(actual.iloc[i:i+5,] )
port_2.Monte_n_var(0.05,1)
```

```python
a=pd.Series(a)
forcast=pd.Series(np.zeros(a.size)+port_2.M_n_var)
BT_M1=bt.Backtest(a,forcast,0.05)
BT_M1.plot()
```

```python
BT_M1.lr_bt()
```

```python
BT_M1.hit_rate()
```
